Your task is to **automatically assign students to fall elective courses** according to their stated preferences and the official algorithm.

#### Official Distribution Algorithm

1. **First Wave**  
   For each course, form a list of all students who indicated it as a **first-priority** choice.  
   - If a student must choose **two** fall electives, both their `fall_1` and `fall_2` choices count as first-priority.  
   If the number of applicants exceeds the course capacity, select the **top students by smallest percentile** (best rating).

2. **Second Wave**  
   For any course that still has free places, form a list of students who indicated it as a **second-priority** choice and fill the remaining spots with the best students (by percentile) from that list.

3. **Third Wave**  
   Repeat the same procedure for **third-priority** choices.

4. Students who still have not received the required number of courses after the three waves will be contacted individually by the academic office.

#### Key Implementation Details

- Competition for each course is **common** for 3rd-year (22nd year) and 4th-year (21st year) students.
- **Default capacity**: every elective course has **1 group of up to 30 students**.
- **Special capacities**:

| Course                                | Capacity |
|---------------------------------------|----------|
| All other courses                     | 30       |
| Quantitative Finance                  | 60       |
| Industrial Programming in Haskell     | 60       |
| Recommender Systems                   | 60       |
| Deep Learning in Audio Processing     | 1000     |

- **Default student load**:
  - Most students choose **1 fall** + **1 spring** elective.
  - 4th-year students also choose **1 blended** course.
- **MOP specialization students** (21st year groups **211, 212, 213** and 22nd-year students with `is_ml_student == True`) must choose **2 fall** + **2 spring** electives.
  - For them, `fall_1` and `fall_2` are both treated as first-priority.
  - Their second-priority is `fall_3`.
  - They participate in only **two selection waves**.

- **Special rule for MOP students and “Machine Learning 2”**:
  - MOP students **cannot** take Machine Learning 2 as a spring elective.
  - If they listed it, their priorities are **shifted**: original priorities 2 and 3 become 1 and 2; the 3rd priority becomes empty.
- In all other cases where a student’s 1st and 2nd choices coincide, **no priority shift** is needed.
- **Blended courses** are **not distributed** in this task and do not affect fall elective assignment.

- **Important assumption** (as stated in the assignment): during distribution there will never be two students with **exactly the same percentile** competing for the last place.

#### Rating and Percentile (how students are ranked)

- The **`rating`** column contains the student’s **credit-rating score** (higher = better).  
- The **`percentile`** column shows the student’s position **within their year and educational program**:
  - **0%** = the best student
  - **100%** = the worst student
- We always select students with the **smallest percentile** (best performers).

````python
def check_first_wave_assignments(assignments: pd.DataFrame) -> None:
    """
    Validates the results of the FIRST WAVE of student course assignments.

    Parameters
    ----------
    assignments : pd.DataFrame
        DataFrame with columns:
            ['ID', 'course1', 'course2', 'fall_course_num']
        One row = one student.
        
        - 'ID'               : student identifier (string)
        - 'course1', 'course2': courses assigned in the first wave.
                                Use None / pd.NA if a course was not assigned.
        - 'fall_course_num'  : number of fall courses the student must choose (1 or 2)

    Raises
    ------
    AssertionError
        If any student's assignment does not match the reference (or allowed alternatives).

    Notes
    -----
    Two specific students have multiple valid outcomes (ALTERNATIVES) due to ties in the algorithm.
    The checker accepts any of the listed possibilities for them.
    """
    
    def course_columns_to_set_column(row: pd.Series) -> Counter:
        """Transform a row's course columns into a Counter of assigned courses."""
        result = Counter()
        for col in ['course1', 'course2']:
            val = row[col]
            if pd.notna(val) and val is not None:
                result[val] += 1
        return result

    ANSWER_SOURCE = '/kaggle/input/datasets/christianmarcos/dataset/task_2_check.csv'

    # Students with multiple acceptable assignments (due to algorithmic ties)
    ALTERNATIVES = {
        "cdb87f7b15495e514eb481230b44540c6b0cfeb347c991f06b1e361fcdef0c0b": [
            Counter({"Deep Learning for Text Data", "Bayesian Methods in Machine Learning"}),
            Counter({"Deep Learning for Text Data"}),
        ],
        "6fa015d8846e21f4f79bac013b469a1e6e962fffd6121b4d9bcde3906275293e": [
            Counter({"Statistical Learning Theory"}),
            Counter({"Statistical Learning Theory", "Bayesian Methods in Machine Learning"}),
        ],
    }

    answer_df = pd.read_csv(ANSWER_SOURCE)
    assignments = assignments.copy()
    assignments['course_set'] = assignments.apply(course_columns_to_set_column, axis=1)
    answer_df['course_set']   = answer_df.apply(course_columns_to_set_column, axis=1)
    mask_common = ~answer_df['ID'].isin(ALTERNATIVES.keys())
    answer_df_common = (
        answer_df[mask_common][['ID', 'course_set', 'fall_course_num']]
        .set_index('ID')
        .sort_index()
    )
    assignments_common = (
        assignments[mask_common][['ID', 'course_set', 'fall_course_num']]
        .set_index('ID')
        .sort_index()
    )
    assert_frame_equal(
        answer_df_common,
        assignments_common,
        check_dtype=False,
        check_names=False,
        check_index_type=False
    )

    # ------------------------------------------------------------------
    # Check students with multiple allowed answers
    # ------------------------------------------------------------------
    for student_id, allowed_sets in ALTERNATIVES.items():
        student_mask = assignments['ID'] == student_id
        
        student_set = assignments.loc[student_mask, 'course_set'].iloc[0]
        assert student_set in allowed_sets, (
            f"Student {student_id} got {student_set} — "
            f"expected one of: {allowed_sets}"
        )

        # fall_course_num must match exactly
        assert (
            answer_df.loc[answer_df['ID'] == student_id, 'fall_course_num'].iloc[0]
            == assignments.loc[student_mask, 'fall_course_num'].iloc[0]
        )

    print("✅ First-wave assignment check PASSED successfully!")
````

#### 3. [2 points] Conduct all three waves of student selection for elective courses


````python
COURSE_CAPACITIES = {
    'Quantitative Finance': 60,
    'Industrial Programming in Haskell': 60,
    'Recommender Systems': 60,
    'Deep Learning for Audio Processing': 1000,
}
DEFAULT_CAPACITY = 30

def get_capacity(course):
    return COURSE_CAPACITIES.get(course, DEFAULT_CAPACITY)


def prepare_priorities(df: pd.DataFrame) -> pd.DataFrame:
    """Prepare p1, p2, p3 priority columns per student."""
    df = df.copy()
    is_mop = df['n_fall_courses'] == 2

    # MOP students with ML2 in any spring slot → shift fall priorities
    has_ml2_spring = (
        (df['spring_1'] == 'Machine Learning 2') |
        (df['spring_2'] == 'Machine Learning 2') |
        (df['spring_3'] == 'Machine Learning 2')
    )
    shift = is_mop & has_ml2_spring

    df['p1'] = df['fall_1']
    df['p2'] = df['fall_2']
    df['p3'] = df['fall_3']

    # Shifted MOP: fall_2→p1, fall_3→p2, p3=''
    df.loc[shift, 'p1'] = df.loc[shift, 'fall_2']
    df.loc[shift, 'p2'] = df.loc[shift, 'fall_3']
    df.loc[shift, 'p3'] = ''

    return df


def distribute_fall(df: pd.DataFrame) -> pd.DataFrame:
    """
    Run 3-wave fall course distribution.
    Returns result DataFrame with columns: ID, course1, course2
    """
    df = prepare_priorities(df).copy()
    is_mop = df['n_fall_courses'] == 2

    # Build capacity tracker
    all_courses = pd.concat([df['p1'], df['p2'], df['p3']]).unique()
    remaining = {c: get_capacity(c) for c in all_courses if c and c != ''}

    # assigned[id] = list of courses assigned
    assigned = {sid: [] for sid in df['id']}

    def fill_course(students_df, priority_col, max_courses):
        """Assign students to their chosen course in priority_col if still needed."""
        for course, group in students_df.groupby(priority_col):
            if not course or course == '':
                continue
            cap = remaining.get(course, 0)
            if cap <= 0:
                continue
            # Eligible: still need more courses
            eligible = group[
                group['id'].map(lambda i: len(assigned[i]) < max_courses)
            ].sort_values('percentile')
            for sid in eligible.head(cap)['id']:
                assigned[sid].append(course)
                remaining[course] -= 1

    mop_df = df[is_mop]
    reg_df = df[~is_mop]

    # --- WAVE 1 ---
    # Regular students: p1 is first priority
    fill_course(reg_df, 'p1', max_courses=1)
    # MOP students: BOTH p1 and p2 are first priority
    fill_course(mop_df, 'p1', max_courses=2)
    fill_course(mop_df, 'p2', max_courses=2)

    # --- WAVE 2 ---
    # Regular students: p2
    fill_course(reg_df, 'p2', max_courses=1)
    # MOP students: p3
    fill_course(mop_df, 'p3', max_courses=2)

    # --- WAVE 3 ---
    # Regular students only: p3
    fill_course(reg_df, 'p3', max_courses=1)

    # Build result DataFrame
    n_courses = dict(zip(df['id'], df['n_fall_courses']))
    rows = []
    for sid, courses in assigned.items():
        if n_courses[sid] == 2:
            course1 = courses[0] if len(courses) > 0 else '???'
            course2 = courses[1] if len(courses) > 1 else '???'
        else:
            course1 = courses[0] if len(courses) > 0 else '???'
            course2 = '-'
        rows.append({'ID': sid, 'course1': course1, 'course2': course2,
                     'fall_course_num': n_courses[sid]})

    return pd.DataFrame(rows).sort_values('ID').reset_index(drop=True)


# Run
result_df = distribute_fall(df)

# Check
check_first_wave_assignments(result_df)

# Save
result_df[['ID', 'course1', 'course2']].to_csv('res_fall.csv', index=False)
print("Done!")
print(result_df.head())
````
